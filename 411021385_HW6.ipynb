```python
from langchain.chains import load_summarize_chain
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.llms import HuggingFacePipeline
from transformers import pipeline

# Load the summarization pipeline
summarization_pipeline = pipeline("summarization", model="google/flan-t5-large")

# Wrap the pipeline in LangChain
llm = HuggingFacePipeline(pipeline=summarization_pipeline)

# Load and read the text file
with open("story.txt", "r") as file:
    text = file.read()
```

```python
# Print the first 1000 characters as a preview
print("Preview of the text:")
print(text[:1000])

# Split the text into chunks
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=100
)
texts = text_splitter.split_text(text)
```

```python
from langchain.docstore.document import Document
```

```python
documents = [Document(page_content=chunk) for chunk in texts]
```

```python
chain = load_summarize_chain(llm, chain_type="map_reduce")
```

```python
summary = chain.run(documents)
```

```python
# Print the final summary
print("\nFinal Summary:")
print(summary)
```
